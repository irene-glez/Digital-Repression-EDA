```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
dend = df_after.drop(columns=['Country', 'Region', 'Regime Type'])
dend.head()
```

```python
# Calculate the distance between each sample
Z = linkage(dend, 'ward')
 
# Plot with Custom leaves
dendrogram(Z, leaf_rotation=90, leaf_font_size=8, labels=dend.index)

# Show the graph
plt.show()
```
